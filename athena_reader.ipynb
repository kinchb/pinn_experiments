```python
%matplotlib widget

import athena_reader
```

```python
athena_data_2 = athena_reader.AthenaReader(
    "../scratch/briowu",
    ("rho", "vel1", "vel2", "vel3", "Bcc1", "Bcc2", "Bcc3", "press"),
)

athena_data_2_5 = athena_reader.AthenaReader(
    "../scratch/briowu_2.5",
    ("rho", "vel1", "vel2", "vel3", "Bcc1", "Bcc2", "Bcc3", "press"),
)
```

```python
t, x = athena_data_2.t, athena_data_2.x
t.shape
```

```python
import matplotlib.pyplot as plt

plt.figure()
t_indices = [0, 50, -1]
colors = ["r", "g", "b"]
for t_ndx, color in zip(t_indices, colors):
    time = t[t_ndx]
    plt.plot(
        x,
        athena_data_2.vel1[t_ndx, :],
        color=color,
        linestyle="-",
        label=f"gamma = 2 (t={time:.2f})",
    )
    plt.plot(
        x,
        athena_data_2_5.vel1[t_ndx, :],
        color=color,
        linestyle="--",
        label=f"gamma = 2_5 (t={time:.2f})",
    )

plt.xlabel("x")
plt.legend()
plt.show()
```
